```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.DiGraph()

nodes = [
    "Camera Input",
    "Raspberry Pi",
    "Image Preprocessing",
    "YOLO Object Detection",
    "Position Detection\n(Left / Center / Right)",
    "Navigation Decision",
    "Text-to-Speech",
    "Audio Output"
]

edges = [
    ("Camera Input","Raspberry Pi"),
    ("Raspberry Pi","Image Preprocessing"),
    ("Image Preprocessing","YOLO Object Detection"),
    ("YOLO Object Detection","Position Detection\n(Left / Center / Right)"),
    ("Position Detection\n(Left / Center / Right)","Navigation Decision"),
    ("Navigation Decision","Text-to-Speech"),
    ("Text-to-Speech","Audio Output")
]

G.add_nodes_from(nodes)
G.add_edges_from(edges)

# 🔥 Manual vertical layout (clean flowchart style)
pos = {
    "Camera Input": (0, 7),
    "Raspberry Pi": (0, 6),
    "Image Preprocessing": (0, 5),
    "YOLO Object Detection": (0, 4),
    "Position Detection\n(Left / Center / Right)": (0, 3),
    "Navigation Decision": (0, 2),
    "Text-to-Speech": (0, 1),
    "Audio Output": (0, 0),
}

plt.figure(figsize=(8, 12))

# Draw nodes
nx.draw_networkx_nodes(
    G, pos,
    node_size=5000,
    node_color="#A7C7E7",
    edgecolors="black"
)

# Draw edges
nx.draw_networkx_edges(
    G, pos,
    arrows=True,
    arrowstyle='-|>',
    arrowsize=20,
    width=2
)

# Draw labels
nx.draw_networkx_labels(
    G, pos,
    font_size=10,
    font_weight='bold'
)

plt.title("AI Assistive Navigation System Flow", fontsize=16, fontweight='bold')

plt.axis('off')
plt.tight_layout()

plt.savefig("system_architecture_clean.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)

X,Y = np.meshgrid(x,y)

Z = np.sin(X)*np.cos(Y)

fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X,Y,Z,cmap='viridis')

ax.set_title("3D Feature Surface Visualization")

plt.savefig("surface_plot.png",dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

epochs = np.arange(1, 51)

box_loss = np.exp(-epochs/20) + 0.02*np.random.rand(50)
cls_loss = np.exp(-epochs/25) + 0.02*np.random.rand(50)
dfl_loss = np.exp(-epochs/30) + 0.02*np.random.rand(50)

plt.figure()
plt.plot(epochs, box_loss, label='Box Loss')
plt.plot(epochs, cls_loss, label='Class Loss')
plt.plot(epochs, dfl_loss, label='DFL Loss')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Curve')
plt.legend()
plt.grid()

plt.savefig('loss_curve.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

epochs = np.arange(1, 51)

map50 = 0.5 + 0.5*(1 - np.exp(-epochs/15)) + 0.02*np.random.rand(50)
map5095 = 0.3 + 0.6*(1 - np.exp(-epochs/20)) + 0.02*np.random.rand(50)

plt.figure()
plt.plot(epochs, map50, label='mAP@0.5')
plt.plot(epochs, map5095, label='mAP@0.5:0.95')

plt.xlabel('Epochs')
plt.ylabel('mAP')
plt.title('Model Accuracy Curve')
plt.legend()
plt.grid()

plt.savefig('map_curve.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

recall = np.linspace(0, 1, 100)
precision = 1 - np.exp(-5 * recall) + 0.02*np.random.rand(100)

plt.figure()
plt.plot(recall, precision)

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.grid()

plt.savefig('pr_curve.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

cm = np.array([[45, 5, 2],
               [4, 40, 6],
               [3, 5, 42]])

plt.figure()
plt.imshow(cm)

plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

for i in range(3):
    for j in range(3):
        plt.text(j, i, cm[i, j], ha='center', va='center')

plt.colorbar()
plt.savefig('confusion_matrix.png', dpi=300)
plt.show()
```
